```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
style.use('ggplot')
```

```python
x,y = np.loadtxt('CSV das Areas.txt', unpack = True, delimiter = ',', usecols = (0,1), dtype = 'object,int')
print(x)
print(y)

''''soma = sum(y)
print(93067*100/soma)
print(70497*100/soma)
print(36668*100/soma)
print(61983*100/soma)
print(15364*100/soma)
print(5573*100/soma)
print(93*100/soma)'''

Soma = sum(y)

Lista_Porcent = []
for i in y:
    c = (i*100/Soma)
    Lista_Porcent.append(int(c))
    
print(Lista_Porcent)
Soma2 = sum(Lista_Porcent)
print(Soma2)

a = False
op = -1

while a == False:
    if Soma2 != 100:
        op += 1
        novo = ((Lista_Porcent[op]) + 1)
        Lista_Porcent[op] = novo
        Soma2 = sum(Lista_Porcent)
    else: 
        a = True
        
print(Lista_Porcent)
print(Soma2)

```
