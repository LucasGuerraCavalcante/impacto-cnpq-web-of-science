```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
x,y = np.loadtxt('Universidade_CAPES_CNPQ.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'object,int')
```

```python
plt.barh(x,y, color = ('gold','peru'))
plt.tick_params(axis='y', which='major', pad=15, length=10)
plt.xticks(color = 'Black')
plt.yticks(color = 'Black')
plt.xlabel('Artigos cofinanciados pela CAPES + CNPq', color = 'Black')

    
for i, v in enumerate(y):
    plt.text(v - 1300, i - 0.2 , str(v), color='Black')
```
