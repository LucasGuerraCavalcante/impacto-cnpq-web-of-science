```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
style.use('ggplot')
```

```python
x,y1 = np.loadtxt('Datas.csv', unpack = True, delimiter = ',', usecols = (0,1), dtype = 'int,int')
print(x)
print(y1)
```

```python
x, y2 = np.loadtxt('BRASIL.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'int,int')
print(x)
print(y2)
```

```python
x, y3 = np.loadtxt('CAPES.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'int,int')
print(x)
print(y3)
```

```python
x, y4 = np.loadtxt('FAPESP.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'int,int')
print(x)
print(y4)
```

```python
plt.plot(x,y1, color = 'blue', marker = 'o')
plt.plot(x,y3, marker = 'o', color = 'c')
plt.plot(x,y4, marker = 'o')
plt.legend(['CNPq','CAPES','FAPESP'])
plt.ylabel('Número de artigos fomentados', color = 'Black')
plt.xlabel('Data de publicação dos artigos', color = 'Black')
plt.xticks(x, color = 'Black')
plt.yticks(color = 'Black')

for i, (a,b,c) in enumerate(zip(y1,y1,x)):
    plt.text(c - 0.8,b + 1000,str(a),color='Black')

plt.show()

#Imapacto do fomento do CNPq no número de publicações de artigos segundo o Web of Science 
```

```python
plt.barh(x,y2, color = 'green')
plt.barh(x,y1, color = 'blue')
plt.ylabel('Ano de Publicação dos Artigos', color = 'Black')
plt.xlabel('Número de Artigos Publicados', color = 'Black')
plt.legend(['Produção Nacional','CNPq'])
plt.xticks(color = 'Black')
plt.yticks(x, color = 'Black')

Lista_Total = []
for i in y2:
    Lista_Total.append(i)
    
    
Lista_CNPq = []
for i in y1:
    Lista_CNPq.append(i)
    
Lista_Porcentagem = []  

for v,a in zip(Lista_Total,Lista_CNPq):
    calculo = (a*100)/v
    Lista_Porcentagem.append(int(calculo))
    
for i, (a,b,c) in enumerate(zip(Lista_Porcentagem,y1,x)):
    plt.text(b + 2500,c - 0.3,'{}%'.format(a),color='Black', ha = 'center')
    

print(Lista_Porcentagem)    
    
plt.show()
    

    
```
