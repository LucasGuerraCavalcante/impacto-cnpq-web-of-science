```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
style.use('ggplot')
```

```python
x,y1 = np.loadtxt('CAPES_Novo.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'int,int')
```

```python
x,y2 = np.loadtxt('FAPESP_CAPES.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'int,int')
```

```python
plt.barh(x,y1, color = 'chocolate')
plt.barh(x,y2, color = 'indianred')
plt.ylabel('Ano de Publicação dos Artigos', color = 'Black')
plt.xlabel('Número de Artigos Publicados', color = 'Black')
plt.legend(['CAPES','CAPES + FAPESP'], loc='lower right')
plt.xticks(color = 'Black')
plt.yticks(x, color = 'Black')

Lista_Total = []
for i in y1:
    Lista_Total.append(i)
    
    
Lista_CNPq = []
for i in y2:
    Lista_CNPq.append(i)
    
Lista_Porcentagem = []  

for v,a in zip(Lista_Total,Lista_CNPq):
    calculo = (a*100)/v
    Lista_Porcentagem.append(int(calculo))
    
for i, (a,b,c) in enumerate(zip(Lista_Porcentagem,y2,x)):
    plt.text(b + 700,c - 0.3,'{}%'.format(a),color='Black', ha = 'center')
```
