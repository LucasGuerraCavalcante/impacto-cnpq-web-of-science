```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
style.use('ggplot')
```

```python
x,y = np.loadtxt('Países.csv', unpack = True, delimiter = ',', usecols = (0,1), dtype = 'object,int')
```

```python
print(x)
print(y)
```

```python
plt.barh(x,y, color=('blue', 'c'))
plt.tick_params(axis='y', pad=10, length=10)
plt.xlabel('Artigos Fomentados Mutuamente com o Brasil (CNPq)', color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.ylabel('Países de filiação de algum dos autores', color = 'Black')

for i, v in enumerate(y):
    plt.text(v + 3, i - 0.2 , str(v), color='Black')

plt.show()

# confirmar
```
