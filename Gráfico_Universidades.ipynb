```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
style.use('ggplot')
```

```python
x,y = np.loadtxt('Organizações.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'object,int')
```

```python
plt.barh(x,y, color = ('Blue','c'))
plt.tick_params(axis='y', which='major', pad=15, length=10)
plt.xticks(color = 'Black')
plt.yticks(color = 'Black')
plt.xlabel('Artigos fomentados pelo CNPq', color = 'Black')

Lista = []
for i in y:
    Lista.append(i)
    
for i, v in enumerate(y):
    plt.text(v + 3, i - 0.2 , str(v), color='Black')



#valores 
```
