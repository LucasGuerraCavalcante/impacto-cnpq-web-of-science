```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
style.use('ggplot')
```

```python
x, y1 = np.loadtxt('Datas.csv', unpack = True, delimiter = ',', usecols = (0,1), dtype = 'int,int')
```

```python
x, y2 = np.loadtxt('BRASIL.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'int,int')
```

```python
x, y3 = np.loadtxt('CAPES.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'int,int')
```

```python
x, y4 = np.loadtxt('FAPESP.csv', unpack = True, delimiter = ';', usecols = (0,1), dtype = 'int,int')
```

```python
plt.plot(x,y1, color = 'blue', marker = 'o')
plt.plot(x,y3, marker = 'o', color = 'c')
plt.plot(x,y4, marker = 'o')
plt.plot(x,y2, color = 'green', marker = 'o')
plt.legend(['CNPq','CAPES','FAPESP','Brasil'])
plt.ylabel('Número de artigos publicados', color = 'Black')
plt.xlabel('Data de publicação dos artigos', color = 'Black')
plt.xticks(x, color = 'Black')
plt.yticks(color = 'Black')


plt.show()
```
